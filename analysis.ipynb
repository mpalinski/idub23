```python
import numpy as np
import pandas as pd
import glob
from collections import Counter
```

```python
all_files = glob.glob("./newspaper/*.csv")

li = []

for filename in all_files:
    try:
        df = pd.read_csv(filename, index_col=None, header=0,lineterminator='\n')
        li.append(df)
    except:
        print(filename)
        pass

df = pd.concat(li, axis=0, ignore_index=True)
```
